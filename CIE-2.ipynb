```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
card_df=pd.read_csv("clean_dataset.csv")
card_df
```

```python
card_df.head()
```

## univariate analysis for continuous variable


```python
sns.histplot(card_df.Age,kde=True)
```

## univariate analysis for categorical variables

```python
sns.countplot(card_df.Gender)
```

## Bivariate analysis for continuous vs continuous variable

```python
sns.scatterplot(card_df.Age,card_df.YearsEmployed)
card_df[['Age','YearsEmployed']].corr()
```

## Bivariate analisis for continuous vs categorical 

```python
sns.kdeplot(data=card_df,x='Age',hue="Approved",fill=True)
```

## Bivariate analysis for categorial vs categorial variable

```python
sns.countplot(data=card_df,x='Gender',hue='Approved')
```

## Multivariate analysis for NUM-NUM- Categorical variable

```python
sns.pairplot(card_df[['Age','Income','YearsEmployed','Gender','CreditScore']],hue='Gender')
```
